#### 4. Catagorical Variable 처리
전처리를 마친 데이터의 변수  
설명변수 X는 `Location`, `Year`, `Kilometers_Driven`, `Fuel_Type`, `Transmission`, `Owner_Type`, `Mileage`, `Engine`, `Power`, `Seats`로 총 10개이고  
반응변수 y는 `Price` 한개입니다.  
이 중, `Location`, `Fuel_Type`, `Transmission`, `Owner_Type` 4개는 범주형 변수입니다.
이제 회귀 분석을 위해 이 4개의 변수를 처리합니다.

```python
# 우선 4변수의 값을 확인합니다.
print(data.Location.unique(), len(data.Location.unique()))
print(data.Fuel_Type.unique(),len(data.Fuel_Type.unique()))
print(data.Transmission.unique(),len(data.Transmission.unique()))
print(data.Owner_Type.unique(),len(data.Owner_Type.unique()))
```

```python
# 4개의 범주형 변수의 데이터를 one-hot으로 바꿔 줍니다.
# 이렇게 바꿔주면 sex별로 두개의 직선을 구할 수 있습니다.
# 변환하여 dummy라고 저장합니다.

dummy_Location = pd.get_dummies(data.Location)
dummy_Location.columns=data.Location.unique()

dummy_Fuel_Type = pd.get_dummies(data.Fuel_Type)
dummy_Fuel_Type.columns=data.Fuel_Type.unique()

dummy_Transmission = pd.get_dummies(data.Transmission)
dummy_Transmission.columns=data.Transmission.unique()

dummy_Owner_Type = pd.get_dummies(data.Owner_Type)
dummy_Owner_Type.columns=data.Owner_Type.unique()



data = pd.concat([data, dummy_Location], axis=1)
data = data.drop(['Location'], axis=1)

data = pd.concat([data, dummy_Fuel_Type], axis=1)
data = data.drop(['Fuel_Type'], axis=1)

data = pd.concat([data, dummy_Transmission], axis=1)
data = data.drop(['Transmission'], axis=1)

data = pd.concat([data, dummy_Owner_Type], axis=1)
data = data.drop(['Owner_Type'], axis=1)

```

```python
# 추가로 필요없는 변수를 제거합니다.
data = data.drop(['Name','Unnamed: 0'], axis=1)

data.info()
```

#### 1. sklearn LinearRegression

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso

from sklearn.metrics import mean_squared_error
```

```python
data.info()
```

```python
X = data.drop(['Price'], axis=1).astype(float)
y = data.Price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 모델을 선언합니다.
model = LinearRegression()

# 모델을 적합합니다.
model.fit(X_train, y_train)
```

```python
# model fitting 결과 입니다.

print(model.intercept_)
print(model.coef_)
```

```python
# 위에서 알 수 있듯이 model은 이미 적합된 모델 객체입니다.
# 따라서 설명변수를 넣어주면 적합된 값을 통해 예측값을 계산해 줍니다.
pred = model.predict(X_train)
print(pred[:3])
```

```python
# fit된 모델의 R-square
model.score(X_train, y_train)
```

```python
mean_squared_error(y_train, pred)
```

```python
# test data R-square
model.score(X_test, y_test)
```

```python
# 예측 vs 실제 데이터 plot
plt.plot(y_test, model.predict(X_test), '.')

# 예측과 실제가 비슷하면, 라인상에 분포함
x = np.linspace(0, 7, 7)
y = x
plt.plot(x, y)
plt.show()
```

이제 다시 앞에서와 같은 방식으로 분석을 진행합니다.

```python
import statsmodels.api as sm
```

```python
X_ = sm.add_constant(X_) 
model = sm.OLS(y, X_)
result = model.fit()
print(result.summary())
```

범주형 데이터를 위주로 제거한 결과 결정계수의 값이 많이 좋아졌고 p-value역시 좋았습니다.

```python
import scipy.stats
```

```python
# 모델을 통해 얻은 예측값 
pred = result.predict()
```

```python
# 잔차 = 실제값 - 예측값 
residual = y - result.predict() 
```

```python
sr = scipy.stats.zscore(residual)
(x, y), _ = scipy.stats.probplot(sr)
```

```python
# 잔차의 정규성 
sns.scatterplot(x, y)
plt.plot([-3, 3], [-3, 3], '--', color='grey')
```

```python
# 잔차의 등분산성 
sns.regplot(pred, np.sqrt(np.abs(sr)), lowess=True, line_kws={'color': 'red'})
```
